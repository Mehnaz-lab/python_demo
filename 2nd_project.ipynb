### **Imputting the Null Values**
-  Imputing null values means replacing missing data with suitable values (like mean, median, mode) so the dataset can be used properly for analysis

```python
df['Recovered'] = df['Recovered'].fillna(df['Recovered'].mean())
df['Deaths'] = df['Deaths'].fillna(df['Deaths'].median())
df['Tests'] = df['Tests'].fillna(df['Tests'].median())   
df['population'] = df['population'].fillna(df['population'].mean())
df['continent'] = df['continent'].fillna(df['continent'].mode()[0])
df.isnull().sum()
```

-------

## Histogram Plot 

```python
sns.histplot(df["continent"], bins=20, kde=True)    
plt.xlabel("Number of Continents")
plt.ylabel("Frequency")
plt.suptitle("Distribution of Continents")
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

### Insight:
- The continent distribution shows how COVID-19 data is spread across different regions, usually reflecting population size, reporting capacity, and overall case impact, with some continents contributing a larger share than others

-------

## Bar plot
  

```python
import matplotlib.pyplot as plt
sns.barplot(x=df['continent'],y=df['Tests'],hue=df['continent'])
plt.xlabel("continent")
plt.ylabel("tests")
plt.xticks(rotation=60)
plt.title("Tests by Continent")  
plt.show()
```

## Insight:
- The **Bar plot** shows that COVID-19 testing varies by continent, with highly populated and developed regions generally conducting more tests than smaller or less resourced continents.

-----

##

## Line Plot 


```python
sns.lineplot(x=df['Deaths'], y=df['Tests'])
plt.xlabel("Deaths")   
plt.suptitle("Deaths vs Tests") 
plt.ylabel("Tests")
plt.grid()
plt.show()
```

## Insight:
- **Line Plot** shows the relationship between COVID-19 deaths and tests conducted.
The line plot suggests that as COVID-19 testing increases, reported deaths may also rise initially due to better detection, but higher testing can help control spread and reduce death rates over time.

------

## Bar plot 

```python
import matplotlib.pyplot as plt
df.groupby('continent')['Recovered'].sum().plot(kind='bar')  
plt.xticks(rotation=45)
plt.xlabel("continent")
plt.ylabel("Cases")
plt.suptitle("Recovered Cases by Continent")
plt.show()
```

## Insight:
- The **Bar plot** indicates that continents with larger populations and higher COVID-19 case numbers, like **Africa** and **Europe**, tend to have more recovered cases, while smaller continents show comparatively lower recoveries.
