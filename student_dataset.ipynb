### **Pie Chart**
- It visualizes the percentage of Pass and Fail students using a pie chart.

```python
df['pass'].value_counts().plot.pie(autopct="%1.1f%%", title="Students Pass/Fail Distribution", labels = ['Pass','Fail'], colors=['lightgreen', 'lightcoral'])
plt.show()
```

------

### **Histograme Chart**
- A Histograme shows how many Students Passed in highest marks & lower marks.

```python
sns.histplot(df['midterm_score'], kde= True, color= "purple")
plt.grid(True)
plt.title("Distribution of Midterm Scores with KDE")
plt.show()
```

------

### **line graph**
- A line graph of study hours per week shows how students’ study time changes over the week.

```python
from statistics import kde
df['study_hours_per_week'].plot(kind='hist',color='teal', bins=15, edgecolor='black')  

title_font = {'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': 16}
label_font = {'family': 'serif', 'color': 'darkred', 'weight': 'bold', 'size': 12}

plt.title('Study Hours Per Week', fontdict=title_font)
plt.xlabel('Index', fontdict=label_font)
plt.ylabel('Study Hours Per Week', fontdict=label_font)
plt.grid(True)
plt.show()
```

------

### Line Plot
- There is a strong positive linear relationship between midterm scores and homework percentage.

```python
sns.lineplot(x= df['midterm_score'], y= df['homework_pct'], color= "teal")

title_font = {'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': 16}
label_font = {'family': 'serif', 'color': 'darkred', 'weight': 'bold', 'size': 12}

plt.grid(True)
plt.title("Midterm Score vs Homework Percentage", fontdict=title_font)
plt.xlabel("Midterm Score", fontdict=label_font)
plt.ylabel("Homework Percentage", fontdict=label_font)
plt.show()
```

### Insights
- There is a strong positive relationship between midterm scores and homework percentage.
As midterm scores increase, homework percentage also increases.
The trend looks linear and consistent.
No major outliers are visible in the graph.

------

### Scatter Plot
- There is a strong positive linear relationship between midterm scores and homework percentage.

```python
sns.scatterplot(x= df['midterm_score'], y= df['homework_pct'], color= "orange")
plt.grid(True)
plt.show()
```

### Insights 
- As midterm scores increase, homework percentage also increases steadily.
The data points follow a clear upward linear trend.
The relationship appears strong and consistent.
No significant outliers are visible in the scatter plot.
